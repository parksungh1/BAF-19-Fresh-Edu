# 19기 신입교육세션 과제1_박성현

#### 데이터명 : Pitchfork Reviews: Music Critiques Over the Years
* 출처 : kaggle (https://www.kaggle.com/datasets/timstafford/pitchfork-reviews)
* 분야 : 음악
* 내용 : 미국의 음악 비평 사이트인 Pitchfork에 게시된 음악 리뷰에 대한 데이터셋입니다. 해당 데이터셋에는 아티스트, 음악(앨범) 발매 년도, 리뷰 작성 일시, 음악 장르와 평점 등의 변수(열)가 포함되어 있습니다. 약 23000개의 곡에 대한 리뷰가 포함되어 있어, 데이터 분석을 진행하는 데 있어 충분히 용이할 것 같아 선정하게 되었습니다. 
* 특히 분석하고 싶은 점은 년도가 최근 년도일수록 같은 음악이어도 평점이 조금 더 후하게 주는 평점 인플레이션이 나타난다고 들었는데 년도에 따라 평점의 증가세가 어떠한 추세를 띄는지 분석해 보고 싶습니다. 

#### 데이터명 : 서울교통공사_역별 일별 시간대별 노인 승하차인원 정보
* 출처 : 서울 열린데이터 광장 (https://data.seoul.go.kr/dataList/OA-22483/F/1/datasetView.do)
* 분야 : 교통
* 내용 : 최근 서울시에서 지하철 노인 무임승차 기준 연령을 65세에서 70세로 상향하면서 이에 대한 사회적 논의가 이어지고 많은 네티즌들이 출근길 교통난 해소를 위해 '출근길에는 노인 무임승차 방안을 폐지하자' 라는 주장을 하는데에 있어 과연 이 주장이 타당한지에 대해 분석을 해볼 수 있을 것 같아 선정하게 되었습니다. 
* 이 데이터셋에서는 서울 주요 지하철역에서 시간대별로 무임승차를 하는 노인들의 수를 날짜별로 다루고 있어 충분한 샘플 수로 분석에 용이합니다.
